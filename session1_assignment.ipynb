```python
# --- Task A: Basic ---

# 1. User Greeting
name = input("Enter your name: ")
age_input = input("Enter your age: ")
age = int(age_input)  # Type casting string to integer

print(f"Hello {name}, you are {age} years old!")

# 2. Basic Arithmetic
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

print(f"{num1} + {num2} = {num1 + num2}")
print(f"{num1} - {num2} = {num1 - num2}")
print(f"{num1} * {num2} = {num1 * num2}")
print(f"{num1} / {num2} = {num1 / num2}")
```

```python
# --- Task B: Intermediate ---

s1 = float(input("Enter score 1: "))
s2 = float(input("Enter score 2: "))
s3 = float(input("Enter score 3: "))

# Calculate average
avg = (s1 + s2 + s3) / 3

# Determine Grade using Comparison & Logical Operators
if avg >= 90:
    grade = "A"
elif avg >= 75:
    grade = "B"
elif avg >= 60:
    grade = "C"
else:
    grade = "F"

# Format output to 2 decimal places using :.2f
print(f"Average score: {avg:.2f}")
print(f"Grade: {grade}")
```

```python
# --- Task C: Advanced ---

raw_input = input("Enter numbers separated by commas: ")
# Split string by comma and convert each element to integer
str_list = raw_input.split(",")
numbers = [int(i.strip()) for i in str_list]

# Initializing variables for manual calculation
total_sum = 0
largest = numbers[0]
smallest = numbers[0]
count = 0

for num in numbers:
    total_sum += num            # Manual Sum
    count += 1                  # Manual Count

    if num > largest:           # Manual Max check
        largest = num

    if num < smallest:          # Manual Min check
        smallest = num

average = total_sum / count

print(f"Sum: {total_sum}")
print(f"Largest: {largest}")
print(f"Smallest: {smallest}")
print(f"Average: {average:.2f}")
```

```python
# --- Bonus Challenge ---

print("\n--- Python Calculator (Type 'exit' to quit) ---")

while True:
    val1 = input("Enter first number (or 'exit'): ")
    if val1.lower() == 'exit': break

    val2 = input("Enter second number: ")
    operator = input("Enter operator (+, -, *, /, %, **): ")

    # Error handling for non-numeric input
    try:
        n1 = float(val1)
        n2 = float(val2)
    except ValueError:
        print("Error: Please enter valid numbers.")
        continue

    # Conditional checks for operations
    if operator == '+':
        print(f"{n1} + {n2} = {n1 + n2}")
    elif operator == '-':
        print(f"{n1} - {n2} = {n1 - n2}")
    elif operator == '*':
        print(f"{n1} * {n2} = {n1 * n2}")
    elif operator == '/':
        if n2 == 0:
            print("Error: Cannot divide by zero")
        else:
            print(f"{n1} / {n2} = {n1 / n2}")
    elif operator == '%':
        print(f"{n1} % {n2} = {n1 % n2}")
    elif operator == '**':
        print(f"{n1} ** {n2} = {n1 ** n2}")
    else:
        print("Invalid operator.")
```
